```python
import pandas as pd
```

```python
python = {'RollNum':[22,33,44,99,66,77,88,99], 'python':[46,78,65,98,67,56,98,78]}
python  = pd.DataFrame(python)
python
```

```python
numpy = {'RollNum':[41,14,53,99,99,34,56,77], 'numpy':[86,66,99,57,66,23,43,54]}
numpy  = pd.DataFrame(numpy)
numpy
```

```python
inner_merged = pd.merge(python,numpy, how='inner')
inner_merged
```

```python
outer_merged = pd.merge(python,numpy, how='outer')
outer_merged
```

```python
right_merge = pd.merge(python,numpy, how='right')
right_merge
```

```python
left_merge = pd.merge(python,numpy, how='left')
left_merge
```
